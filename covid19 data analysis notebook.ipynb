### Task 2.1: importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
file_path = "./Datasets/covid19_Confirmed_dataset.csv"
covid_df = pd.read_csv(file_path)
covid_df.head(10)
```

#### Let's check the shape of the dataframe

```python
covid_df.shape
```

### Task 2.2: Delete the useless columns

```python
covid_df.drop(['Lat', "Long"], axis=1, inplace=True)
```

```python
covid_df.head()
```

### Task 2.3: Aggregating the rows by the country

```python
covid_aggregated = covid_df.groupby(['Country/Region']).sum()
```

```python
covid_aggregated.head()
```

```python
covid_aggregated.shape
```

### Task 2.4: Visualizing data related to a country for example China

### task 3.1: caculating the first derivative of the curve

```python
covid_aggregated.loc["China"].diff().plot();
```

### task 3.2: find maxmimum infection rate for China

```python
covid_aggregated.loc["China"].diff().max()
```

```python
covid_aggregated.loc["Italy"].diff().max()
```

```python
covid_aggregated.loc["Spain"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(covid_aggregated.index)
max_infection_rates = []
for country in countries:
    max_infection_rates.append(covid_aggregated.loc[country].diff().max())
```

```python
covid_aggregated["max_infection_rate"] = max_infection_rates
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(covid_aggregated["max_infection_rate"])
```

```python
corona_data
```
